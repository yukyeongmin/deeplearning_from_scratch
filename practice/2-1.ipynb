```python
import numpy as np
```

```python
class Affine:
    def __init__(self,W,b): # weight, bias
        self.params=[W,b]
        
    def forward(self,x):
        W,b = self.params
        out = np.matmul(x,W)+b
        return out
    
        
class Sigmoid:
    def __init__(self):
        self.params=[]
        
    def forward(self,x):
        return 1/(1+np.exp(-x))
```

```python
class TwolayerNet:
    def __init__(self,input_size,hidden_size,output_size):
        I, H, O=input_size, hidden_size, output_size
        
        # 첫번째 계층의 weight, bias
        W1=np.random.randn(I,H)
        b1=np.random.randn(H)
        # 두번째 계층의 weight, bias
        W2=np.random.randn(H,O)
        b2=np.random.randn(O)
        
        # 계층 생성
        self.layers=[
            Affine(W1,b1),
            Sigmoid(),
            Affine(W2,b2)
        ]
        
        self.params=[]
        for layer in self.layers:
            self.params+=layer.params
            
    def predict(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x
```

```python
x = np.random.randn(10,2)
model = TwolayerNet(2,4,3)
s = model.predict(x)
```

```python
import pandas

data=np.stack((x[:,0],x[:,1],s[:,0],s[:,1],s[:,2]),axis=-1)
headers=['x1','x2','label1','label2','label3']
pandas.DataFrame(data,columns=headers)
```
